```python
import pandas as pd
import plotly.express as px
```

```python
test = pd.read_csv('all_samples_23_6.csv')
test['max_profit']= (test['max_price'] - test['open_price'])/test['open_price']
test['min_profit']= (test['min_price'] - test['open_price'])/test['open_price']
test['dt_stamp'] = pd.to_datetime(test['open_trade_dt'])
test['dow'] = test['dt_stamp'].dt.dayofweek
print(len(test))
```
